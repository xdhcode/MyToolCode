```python
import os
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib notebook
from sklearn.preprocessing import MinMaxScaler
from tqdm import tqdm
#准备文件夹保存数据
local_path=os.getcwd()
if not os.path.exists(local_path+'\\blokagedata'):
    os.mkdir(local_path+'\\blokagedata')
print(local_path+'\\blokagedata')

#读取模型数据
sample=pd.read_csv(local_path+'\\blokagedata\\Sample_116.csv',header=0)
sample_p=pd.read_csv(local_path+'\\blokagedata\\Sample_116_percent.csv',header=0)
def find_code_time(sample,code,time):
    label=np.where(sample['code']==code)[0]
    temp=sample.iloc[label,:]
    label=np.where(temp['time']==time)[0]
    temp=temp.iloc[label,:]
    return temp
def find_dn_time(sample,dn,time):
    label=np.where(sample['dn']==dn)[0]
    temp=sample.iloc[label,:]
    label=np.where(temp['time']==time)[0]
    temp=temp.iloc[label,:]
    return temp
def find_code_dn(sample,code,dn):
    label=np.where(sample['code']==code)[0]
    temp=sample.iloc[label,:]
    label=np.where(temp['dn']==dn)[0]
    temp=temp.iloc[label,:]
    return temp
```

```python
for i in range(0,24):
    label=np.where(sample['time']==i)[0]
    temp=sample.iloc[label,:].T
    temp.to_csv(local_path+'\\blokagedata\\Sample_time_'+str(i)+'_percent.csv',encoding='ascii',index=True)
```

# dn

```python
#比较dn的影响
code=36
time=0
compare=find_code_time(sample=sample,code=code,time=time)

z=0
x=range(z,96)
plt.figure(figsize=(9,9),dpi=100)
for i in range(0,len(compare)):
    y=compare.iloc[i,range(z,96)]
    plt.plot(x,y,marker='.',label=('dn:',compare.iloc[i,97]))
plt.title('Residual_dn')
plt.xlabel('user id')
plt.ylabel('Residual')
plt.legend()
plt.show()
```
